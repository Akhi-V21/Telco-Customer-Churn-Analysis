## Task 2 & 6: Exploratory Data Analysis & Advanced Visualizations

### Objective:
In this section, I explore the dataset to identify patterns that correlate with customer churn. Using advanced visualizations, we can pinpoint which customer segments are most likely to leave the service.

### Analysis Goals:
1. **Target Distribution:** Check the balance of Churn (Yes vs. No).
2. **Contract Type Impact:** Visualize how contract duration influences loyalty.
3. **Monthly Charges & Tenure:** Use box plots and violin plots to see the financial and time-based distribution of churned customers.

```python
import pandas as pd

# Load the Telco Customer Churn dataset from a public repository
url = 'https://raw.githubusercontent.com/IBM/telco-customer-churn-on-icp4d/master/data/Telco-Customer-Churn.csv'
df = pd.read_csv(url)

# Display the first 5 rows to ensure it loaded correctly
display(df.head())
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Set a professional style
sns.set_theme(style="whitegrid")
plt.rcParams['figure.figsize'] = (12, 6)

# 1. Churn Pie Chart (Overall View)
plt.figure(figsize=(6, 6))
df['Churn'].value_counts().plot.pie(autopct='%1.1f%%', colors=['#66b3ff','#ff9999'], startangle=90)
plt.title('Percentage of Customer Churn')
plt.ylabel('')
plt.show()

# 2. Churn by Contract Type (Task 2 Insight)
plt.figure(figsize=(10, 5))
sns.countplot(x='Contract', hue='Churn', data=df, palette='viridis')
plt.title('Churn Count by Contract Type')
plt.show()

# 3. Monthly Charges Distribution - Violin Plot (Task 6 Requirement)
plt.figure(figsize=(10, 6))
sns.violinplot(x='Churn', y='MonthlyCharges', data=df, inner="quart", palette='Set2')
plt.title('Distribution of Monthly Charges by Churn')
plt.show()

# 4. Tenure vs Churn - Box Plot (Task 6 Requirement)
plt.figure(figsize=(10, 6))
sns.boxplot(x='Churn', y='tenure', data=df, palette='Set1')
plt.title('Tenure (Months) vs Churn Status')
plt.show()
```
